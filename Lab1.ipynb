## NumPy (створити набір даних (розмір за власним рішенням), провести CRUD операції над даними (скаляр, вектор, матриця)). Властивість shape та метод reshape. Операції над масивами. Використати arange, random.

```python
import numpy as np
import random as rand

a=1
b=2
c=3

mas = np.array([
                [a, b, c], 
                [b, a, c], 
                [b, c, a], 
                [c, a, b]]) #створення масиву

print(np.append(mas, [[3, 4, 5]], axis=0)) # вставка

print(mas.shape[1]) #розмірність на другому рівні
print(mas.reshape((3, 4))) #перетворення масиву

print(np.arange(0, 5, 0.5)) #створення масиву від до з кроком

print(rand.random()) #випадкове число від 0 до 1
```
